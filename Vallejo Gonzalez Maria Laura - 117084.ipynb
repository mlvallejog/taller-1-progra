```python
'''El programa va a contar :
    a) cuántas letras vocales en mayúscula se entraron;
    b) cuántas letras con tilde se entraron (minúsculas y mayúsculas);
    c) cuántos dígitos se entraron, 
    d) cuántos espacios se entraron;
    e) cuántas palabras reservadas se entraron.
 dentro de un texto ingresado
'''


#pedir que se ingrese la cadena de texto

cadena = input ('por favor ingrese un texto : ')

#convertir los datos ingresados en una lista 
cadena_lista =cadena.split() #split() agrega la cadena a una lista como elementos seperados

#identificar los carácteres que pide el ejercicio

numeros             =  '0123456789'
letras_con_tilde    =  'ÁÉÍÓÚáéíóú'
vocal_en_mayus      =  'AEIOU'
espacio             =  ' '

palabras_reservadas =['False','None','True','and','as','assert','break','class','continue','def','del','elif','else',\
                      'except','finally','for','from','global','if','import','in','is','lambda','nonlocal','not','or',\
                      'pass','raise','return','try','while','with','yield']

#crear variables donde se va a empezar a sumar los encontrados

contador_de_voc_mayus       =0

contador_de_letras_con_til  =0

contador_de_digitos         =0

contador_de_espacios        =0

contador_de_pal_reser       =0 



#ciclo identificar y sumar cada variable pedida

for i in range(len( cadena )):
    
    
    if cadena[i] in vocal_en_mayus: #contador de vocales mayus
         contador_de_voc_mayus += 1
   

    elif cadena[i] in tildes:  #contador de letras con tilde
        contador_de_letras_con_til += 1
    
    
    
    elif cadena[i] in digitos: #contador de digitos
        contador_de_digitos +=1
    
    
    elif cadena[i]==espacio:  #contador de espacios
        contador_de_espacios +=1

        
        
#el contador de palabras reservadas debe obedecer a buscar en una lista en vez de en una cadena
         

    
    
    
    for j,i in enumerate (cadena_lista) :    #iterar en los elementos de la lista y comparar
       
        if i in palabras_reservadas:          
               contador_de_pal_reser+=1      

                
                
#imprimir todo
        
print(f'la cadena tiene {contador_de_voc_mayus}  vocales en mayúscula')
print(f'la cadena tiene {contador_de_letras_con_til} letras con tilde')
print(f'la cadena tiene {contador_de_digitos} digitos ó números')
print(f'la cadena tiene {contador_de_espacios} espacios')
print(f'la cadena tiene {contador_de_pal_reser} palabras reservadas')
```

```python

import random as rd
def Suma():
    
    '''  El programa genera sumas con numeros aleatorios del 1 al 100, el usuario debe acertar en 
        la respuesta y seguirá hasta que el usuario acierte 5 sumas.
    '''
    #crear variable para contar aciertos 
    acertar = 0
    
    
    
    #ciclo para preguntar mientras no se acierte 
    while  acertar < 5:   
        num1  = rd.randint (1,100)
        num2   = rd.randint (1,100) 
        sumado = num1 + num2 
        print ('\nescriba el resultado de las siguientes operaciones ')
        print ('\n',num1 ,'+',num2,'=')
        #print ( sumado )    #para verificar el programa
        almacenar = int(input('la suma es =  '))  #pedir la respuesta de la suma
        
        #si la respuesta es correcta sumar un intento correcto
        if almacenar == sumado : 
                 print ('\nRespuesta Correcta !')
                 acertar += 1
        else: print('\nVuelve a intentarlo')
     
        
    print( '\nfelicitaciones has completado los 5 aciertos !')
        

```
